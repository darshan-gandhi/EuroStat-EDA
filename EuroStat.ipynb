Q2 Find the mean value of all the all pollutions in BG over the years

```python
new_data=data[(data.geotime=='BG')]
```

```python
new_data
```

```python
new_data=new_data.drop(columns={'essentials','unit','src_nfr','airpollution','geotime'})
```

```python
new_data=new_data.astype("float64")
answers=list(new_data.sum(axis=1))
store=[]
for i in answers:
    ans=float(i/27)
    store.append(ans)
store
```

```python
list1=data['airpollution'].unique()
df_mean=pd.DataFrame(list1,columns=['Airpollution'])
df_mean['Mean']=store
```

```python
df_mean
```

Q4 Find the top5 country having the highest value for the emission of PM10 in the year 2002

```python
save=data[['2002','geotime']][(data.airpollution=='PM10')]
save
```

```python
save['2002']=save.astype({"2002":"float64"})
save.sort_values(by=['2002'],ascending=False).head(5).plot.bar(x='geotime',y='2002')
```
